```python
# Exploratory data analysis
```

```python
sns.set(style="whitegrid")
for feature in data.columns:
    if data[feature].dtypes in (["object"]):
        data[feature].value_counts().plot(kind="bar",colors=list("rgbkymc"))
        plt.xlabel(feature)
        plt.ylabel("Count")
        plt.title(feature)
        plt.show()
```

```python
# Male vs female
```

```python
# lets analyze who is better in all subjects
```

```python
male=data[data["gender"]=="male"]
female=data[data["gender"]=="female"]
```

```python
# score of male students
```

```python
result=[]
subject=[]
for feature in male.columns:
    if male[feature].dtypes in (["int64"]):
        result.append(round(male[feature].sum()/len(male),3))
        subject.append(feature)
        sns.barplot(x=subject,y=result)
        plt.ylabel("mean marks")
        plt.xlabel("subjects")
        plt.title("Male students marks")
```

```python
# score of female students 
```

```python
result=[]
subject=[]
for feature in female.columns:
    if female[feature].dtypes in (["int64"]):
        result.append(round(female[feature].sum()/len(female),3))
        subject.append(feature)
        sns.barplot(x=subject,y=result)
        plt.ylabel("mean marks")
        plt.xlabel("subjects")
        plt.title("female students marks")
```

```python
# we can easily find out in which subject boys are good and in which girls are good
```

```python
# Data visualization
```

```python
for feature in data.columns:
    if data[feature].dtypes in (["object"]):
        sns.countplot(data[feature])
        plt.xlabel(feature)
        plt.title(feature)
        plt.xticks(rotation=90)
        plt.show()
```

```python
# Lets see it gender wise
```

```python
for feature in data.columns:
    if data[feature].dtypes in (["object"]):
        sns.countplot(data[feature],hue="gender",data=data,palette="Set2")
        plt.xlabel(feature)
        plt.show()
```

```python
# Inference
```

```python
'''
What can we get from the information above? 3 out of 5 groups, have the most male. And we get some insights here:

Group C has more differences in the number of male and female than the others
The group with the least number of male is group A, and so are the female
The group with the most number of female is group C, and so are the male

'''
```

```python
# swarm plot
```

```python
fig=plt.figure(figsize=(15,5))
for feature in data.columns:
    if data[feature].dtypes in (["int64"]):
        sns.swarmplot(data[feature],y="gender",data=data,palette="Set2")
        plt.xlabel(feature)
        plt.title(feature)
        plt.show()
```

```python
# by looking at the last graph we can say that females lead males in getting a score 
# of 100 and also leads in getting least marks
```

```python
# Pie plot
```

```python
'''
Let's assume if there are students who score < 60, then we determine they did not pass the exam.
'''
```

```python
#math score
passed = len(data[data['math score'] >= 60])
not_passed = 1000 - passed

percentage1 = [passed, not_passed]

#reading score
passed = len(data[data['reading score'] >= 60])
not_passed = 1000 - passed

percentage2 = [passed, not_passed]

#writing score
passed = len(data[data['writing score'] >= 60])
not_passed = 1000 - passed

percentage3 = [passed, not_passed]
```

```python
labels = "Passed", "Not Passed"

f, axs = plt.subplots(2,2, figsize=(15,10))

#Math Score
axs[0,0].pie(percentage1, labels = labels, explode=(0.05,0.05), autopct = '%1.1f%%', startangle = 200, colors = ["#e056fd", "#badc58"])
axs[0,0].set_title("Math Score", size = 20)
axs[0,0].axis('equal')

#Reading Score
axs[0,1].pie(percentage2, labels = labels, explode=(0.05,0.05), autopct = '%1.1f%%', startangle = 200, colors = ["#e056fd", "#badc58"])
axs[0,1].set_title("Reading Score", size = 20)
axs[0,1].axis('equal')

#Writing Score
axs[1,0].pie(percentage3, labels = labels, explode=(0.05,0.05), autopct = '%1.1f%%', startangle = 200, colors = ["#e056fd", "#badc58"])
axs[1,0].set_title("Writing Score", size = 20)
axs[1,0].axis('equal')

f.delaxes(axs[1,1]) #deleting axs[1,1] so it will be white blank

plt.show()
```

```python
'''
We get some information:

323 students didn't pass Math exam
254 students didn't pass Reading exam
281 students didn't pass Writing exam
'''
```

```python
'''
A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.
'''
```
